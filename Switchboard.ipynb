```python
mocDict = dict(zip(mocs, range(len(mocs))))
for index in range(switchboard.shape[0]):
    moc = switchboard.act_label_1.iat[index]
    switchboard.act_label_1.iat[index] = mocDict[moc]
```

```python
from IPython.display import clear_output
from joblib import Parallel, delayed
import multiprocessing
num_cores = multiprocessing.cpu_count() // 2

documents = []

def convertConversation(group):
    document = {}
    document['convo'] = []
    turns = []
    lastCaller = None
    lastMOC = None
    for index in range(group.shape[0]):
        text = group.clean_text.iat[index]
        caller = group.caller.iat[index]
        moc = group.act_label_1.iat[index]
        if lastCaller == caller and lastMOC == moc:
            turns[-1][0] += " " + text
        else:
            turns.append([text, caller, moc])
        lastCaller = caller
        lastMOC = moc
    for text, caller, moc in turns:
        try:
            sentences = cn.basic(text).json()['sentences']
        except Exception as e:
            print(e)
            print(text)
        endOfLast = 0
        for sent in sentences:
            last_word = sent['tokens'][-1]
            sentence_text = text[endOfLast:last_word['characterOffsetEnd']]
            endOfLast = last_word['characterOffsetEnd']
            tokens = list(map(lambda x: x['word'], sent['tokens']))
            document['convo'].append({'speaker': caller, 'text': sentence_text, 'tokens': tokens, 'moc': moc})
    documents.append(document)
for name, group in tqdm(switchboard.groupby('conversation_no')):
    convertConversation(group)
# results = Parallel(n_jobs=num_cores)(delayed(convertConversation)(group) for name, group in switchboard.groupby('conversation_no'))
```

```python
model = gensim.models.Word2Vec.load_word2vec_format('./datasets/google/word2vec.bin', binary=True)
```

```python
def turnToVec(turn):
    vec = np.zeros((300,))
    for token in turn['tokens']:
        try:
            vec += model[token]
        except KeyError:
            pass
    return vec
```

```python
X = []
y = []
for doc in tqdm(documents):
    for turn in doc['convo']:
        X.append(turnToVec(turn))
        y.append(turn['moc'])
```
